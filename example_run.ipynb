### 1. Import the main class from the lfploc library

```python
from lfploc import Lfploc
```

### 2. Initialize your recording with SpikeInterface and ensure a probe is linked

```python
rec_path = r"E:\recording.h5"


from spikeinterface.extractors.sinapsrecordingextractors import read_sinaps_research_platform_h5
rec = read_sinaps_research_platform_h5(rec_path)

assert rec.has_probe(), "Probe not attached. Link one with rec.set_probe(probe, in_place=True)"
```

### 3. Initialize the Lfploc class and ensure it is configured correctly by validating the probe's geometry specifications

```python
loc = Lfploc(rec)

print(loc.probe_specs)
```

### 3.1. If something is not correct simply change it by modifying the loc.probe_specs object

```python
loc.probe_specs["tip_angle"] = 25
```

### 3.2. Find out more about all the parameters passed to run

```python
print(loc.get_run_arguments_info())
```

### 4. Run the full analysis

```python
cluster_labels = loc.run(
    feature_extraction_method="PCA",
    clustering_method="k-means",
    end_time=20,
    save_report_dir="lfploc_report",
    ap=-2.3,
    ml=2.4,
    dv=-2
)
# this command will generate plots in the "save_report_dir" directory, if you do not want plots generated and just the cluster_labels don't pass the "save_report_dir" argument
```

### 4a. EXTRA: Plot only the stereotactic coordinates on the atlas without probe

```python
loc.place_probe_on_atlas(
    ap=-2.3,
    ml=2.4,
    dv=-2,
    save_report_dir=r"only_coordinates_lfploc"
)
```
